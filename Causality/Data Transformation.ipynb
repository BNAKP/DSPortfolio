```python
rawData.dropna(axis=1,how='all',inplace=True)
```

## Pivot Data

```python
disagg=rawData.columns[4:].tolist()
col=['Indicator','SeriesCode']
for j in range(0,len(disagg)):
    col.append(disagg[j])
```

```python
col
```

```python
Data=rawData.pivot(index='TimePeriod',columns=col,values='Value')
```

```python
Data
```

### Drop columns with less than 3 values (correlation only made with indicators with at least 3 values)

```python
Data.dropna(axis=1,thresh=3,inplace=True)
```

## Create new dataframe

```python
#Getting years from original data
Years=Data.index.values
```

```python
#Creating first column of new dataframe (to check code works and makes it easier to add new columns)
col1=Data.iloc[:,0]
col1val=col1.values
col1name=col1.name
col1code=col1name[1]+'--'

for n in range(len(col1name)-2):
    if type(col1name[n+2]) == str:
        col1code=col1code+col1name[n+2]+'_'
        
        
cleanData=pd.DataFrame(col1val,index=Years,columns=[[col1name[0]],[col1code]])
```

```python
cleanData
```

```python
#Adding remaining columns to dataframe
for k in range(1,len(Data.columns)):
    colname=Data.iloc[:,k].name
    colcode=colname[1]+'--'
    for n in range(2,len(colname)):
        if type(colname[n]) == str:
            colcode=colcode+colname[n]+'_'
    cleanData[colname[0],colcode]=Data.iloc[:,k].values
```

```python
cleanData
```

## Perform correlation

```python
corrmat=cleanData.corr(method="spearman",min_periods=3)
```
